```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
protocol = pd.read_csv("protocol_KOPP_versuch_updated_27_april22.csv")
```

```python
protocol.drop([7,11], axis=0, inplace=True)
```

```python
protocol = protocol.sort_values(by=['Number of usages'])
```

```python
protocol
```

```python
protocol3 = protocol[protocol['Tool ID']== '3']
```

```python
protocol3
```

```python
durchmesser3 = protocol3['Durchmesser']
span3 =  protocol3['Spanwinkel']
rund3 =  protocol3['Rundlauf']
radial3 =  protocol3['Radialwinkel']
uses3 = protocol3['Number of usages']

plt.figure(figsize = (12,6))
plt.title("TOOL 3 Protocol")
plt.subplot(2,2,1)
plt.plot(uses3, durchmesser3)
plt.scatter(uses3, durchmesser3, label = 'durchmesser', color = 'r')
plt.legend()
plt.xlabel("No of usages")


plt.subplot(2,2,2)
plt.plot(uses3, span3)
plt.scatter(uses3, span3, label = 'spanwinkel', color = 'r')
plt.legend()
plt.xlabel("No of usages")


plt.subplot(2,2,3)
plt.plot(uses3, rund3)
plt.scatter(uses3, rund3, label = 'rundlauf', color = 'r')
plt.legend()
plt.xlabel("No of usages")


plt.subplot(2,2,4)
plt.plot(uses3, radial3)
plt.scatter(uses3, radial3, label = 'radialwinkel', color = 'r')
plt.legend()
plt.xlabel("No of usages")

plt.show()

####  TOOL 3 protocol values  ####
```

```python
plt.figure(figsize = (12,6))
plt.plot(protocol['Number of usages'], protocol['Durchmesser'] )
plt.scatter(protocol['Number of usages'], protocol['Durchmesser'], color = 'r' )
plt.title("Durchmesser_SR_process")
plt.xlabel("No of usages")
plt.ylabel("Durchmesser")


uses = protocol['Number of usages']
durchmesser = protocol['Durchmesser']
```

```python
trend3 = np.polyfit(uses, durchmesser, 3)
trend2 = np.polyfit(uses, durchmesser, 2)
trend1 = np.polyfit(uses, durchmesser, 1)

plt.figure(figsize = (15,6))
x = np.arange(0,30,1)
plt.plot(uses, durchmesser,'o')
plt.plot(uses, durchmesser)
trendpoly1 = np.poly1d(trend1) 
plt.plot(x,trendpoly1(x), label = 'Degree 1')
plt.legend()
plt.show()

plt.figure(figsize = (15,6))
plt.plot(uses, durchmesser,'o')
plt.plot(uses, durchmesser)
trendpoly2 = np.poly1d(trend2) 
plt.plot(x,trendpoly2(x), label = 'Degree 2')
plt.legend()
plt.show()

plt.figure(figsize = (15,6))
plt.plot(uses, durchmesser,'o')
plt.plot(uses, durchmesser)
trendpoly3 = np.poly1d(trend3) 
plt.plot(x,trendpoly3(x), label = 'Degree 3')
plt.legend()
plt.show()

plt.figure(figsize = (15,6))
plt.plot(uses, durchmesser,'o')
plt.plot(uses, durchmesser)

plt.plot(x,trendpoly1(x), label = 'Degree 1')
plt.plot(x,trendpoly2(x), label = 'Degree 2')
plt.plot(x,trendpoly3(x), label = 'Degree 3')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (12,6))
plt.plot(protocol['Number of usages'], protocol['Spanwinkel'] )
plt.scatter(protocol['Number of usages'], protocol['Spanwinkel'], color = 'r' )
plt.title("Spanwinkel_SR_process")
plt.xlabel("No of usages")
plt.ylabel("Spanwinkel")

plt.show()
spanwinkel = protocol['Spanwinkel']

x = np.arange(0,30,1)

plt.figure(figsize = (15,6))
trend = np.polyfit(uses, spanwinkel, 1)
plt.plot(uses, spanwinkel,'o')
plt.plot(uses, spanwinkel)
trendpoly = np.poly1d(trend) 
plt.plot(x,trendpoly(x), label = 'Degree 1')
#plt.plot(x,trendpoly(x))
plt.show()

plt.figure(figsize = (15,6))
trend2 = np.polyfit(uses, spanwinkel, 2)
plt.figure(figsize = (15,6))
plt.plot(uses, spanwinkel,'o')
plt.plot(uses, spanwinkel)
trendpoly2 = np.poly1d(trend2)
plt.plot(x,trendpoly2(x), label = 'Degree 2')
plt.legend()
plt.show()
```
