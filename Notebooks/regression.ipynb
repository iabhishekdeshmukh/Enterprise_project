## Linear regression using np.polyfit
### 1 D feature

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Data

```python
protocol = pd.read_csv("protocol_KOPP_versuch_updated_27_april22.csv")
protocol.drop([7,11], axis=0, inplace=True)
protocol = protocol.sort_values(by=['Number of usages'])
protocol3 = protocol[protocol['Tool ID']== '3']
```

## Regression 

```python
degree = 1                           #degree for polynomial fitting
x = protocol3['Number of usages']    #x axis for regression
y = protocol3['Durchmesser']         #y axis valus the regression function to fit with  
```

```python
def regression1(x, y, degree):
    trend = np.polyfit(x, y, degree)
    trendpoly = np.poly1d(trend) 
    return trendpoly             #return regression function
```

```python
reg_func = regression1(x, y, degree)
plt.plot(x,reg_func(x), label = 'regression') #plot regression
plt.plot(x, y, label = "original function")   #plot original values 
plt.legend()
plt.show()
```
